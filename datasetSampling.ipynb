```python
import csv
import numpy as np
```

```python
# make the csv data into an array
data = []
with open("dataset.csv") as file:
    reader = csv.reader(file)
    for row in reader:
        # don't read in song number or track_id
        data.append(row[2:])
```

```python
r = len(data)
# choose 1000 random songs that we want to use as our dataset
indices = np.random.choice(np.arange(r), 3000, replace=False)
data3000 = [data[i] for i in indices]
```

```python
print(data3000)
# artistname, trackname, playlistname
```

```python
# write the 1000 songs to a new csv
output_file = "data3000.csv"
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data1000)
```
